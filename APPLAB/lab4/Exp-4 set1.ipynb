```python
#Q9
'''Note: This is a very simple calculator with many limitations.
It can perform simple operations only on single digit positive numbers'''
from tkinter import *
root=Tk()
root.title('9. Calculator')
root.configure(bg='orange')
text=StringVar()
listval=[]
def click_1():
    text.set('1')
    listval.append(1)
def click_2():
    text.set('2')
    listval.append(2)
def click_3():
    text.set('3')
    listval.append(3)
def click_4():
    text.set('4')
    listval.append(4)
def click_5():
    text.set('5')
    listval.append(5)
def click_6():
    text.set('6')
    listval.append(6)
def click_7():
    text.set('7')
    listval.append(7)
def click_8():
    text.set('8')
    listval.append(8)
def click_9():
    text.set('9')
    listval.append(8)
def click_0():
    text.set('0')
    listval.append(9)
    
def click_plus():
    text.set('+')
    listval.append('+')
def click_minus():
    text.set('-')
    listval.append('-')
def click_div():
    text.set('/')
    listval.append('/')
def click_mult():
    text.set('*')
    listval.append('*')
def click_pow():
    text.set('^')
    listval.append('^') 
def click_mod():
    text.set('%')
    listval.append('%') 

     

def click_result():
    ans=0
    if(listval[1]=="+"):
        ans+=(listval[0]+listval[2])
    elif(listval[1]=="-"):
        ans+=(listval[0]-listval[2])
    elif(listval[1]=="/"):
        ans+=(listval[0]/listval[2])
    elif(listval[1]=="*"):
        ans+=(listval[0]*listval[2])
    elif(listval[1]=="^"):
        ans+=(listval[0]**listval[2])
    elif(listval[1]=="%"):
        ans+=(listval[0]%listval[2])
    
        
    for i in range(3,len(listval)):
        if(listval[i]=='+'):
            ans+=listval[i+1]
        elif(listval[i]=='-'):
            ans-=listval[i+1]
        elif(listval[i]=='*'):
            ans*=listval[i+1]
        elif(listval[i]=='/'):
            ans/=listval[i+1]
        elif(listval[i]=='^'):
            ans**=listval[i+1]
        elif(listval[i]=='%'):
            ans%=listval[i+1]
        
    text.set(str(ans))

CalculatorDisplay=Label(root,textvariable=text,width=15,height=1,bg='light blue').grid(row=0,column=1)
le=Label(root,bg="orange").grid(row=1,column=0)
b1=Button(root,text='1',command=click_1,width=5,height=3,bg='light green',activebackground='red').grid(row=2,column=0)
b2=Button(root,text='2',command=click_2,width=5,height=3,bg='light green',activebackground='red').grid(row=2,column=1)
b3=Button(root,text='3',command=click_3,width=5,height=3,bg='light green',activebackground='red').grid(row=3,column=0)
b4=Button(root,text='4',command=click_4,width=5,height=3,bg='light green',activebackground='red').grid(row=3,column=1)
b5=Button(root,text='5',command=click_5,width=5,height=3,bg='light green',activebackground='red').grid(row=4,column=0)
b6=Button(root,text='6',command=click_6,width=5,height=3,bg='light green',activebackground='red').grid(row=4,column=1)
b7=Button(root,text='7',command=click_7,width=5,height=3,bg='light green',activebackground='red').grid(row=5,column=0)
b8=Button(root,text='8',command=click_8,width=5,height=3,bg='light green',activebackground='red').grid(row=5,column=1)
b9=Button(root,text='9',command=click_9,width=5,height=3,bg='light green',activebackground='red').grid(row=6,column=0)
b0=Button(root,text='0',command=click_0,width=5,height=3,bg='light green',activebackground='red').grid(row=6,column=1)

b_plus=Button(root,text='+',command=click_plus,width=5,height=3,bg='light green',activebackground='red').grid(row=2,column=2)
b_minus=Button(root,text='-',command=click_minus,width=5,height=3,bg='light green',activebackground='red').grid(row=3,column=2)
b_div=Button(root,text='/',command=click_div,width=5,height=3,bg='light green',activebackground='red').grid(row=4,column=2)
b_mult=Button(root,text='*',command=click_mult,width=5,height=3,bg='light green',activebackground='red').grid(row=5,column=2)
b_equal=Button(root,text='=',command=click_result,width=5,height=3,bg='light green',activebackground='red').grid(row=6,column=2)
b_pow=Button(root,text='^',command=click_pow,width=5,height=3,bg='light green',activebackground='red').grid(row=7,column=2)
b_mod=Button(root,text='%',command=click_mod,width=5,height=3,bg='light green',activebackground='red').grid(row=7,column=0)
b_close=Button(root,text='X',command=root.destroy,width=5,height=3,bg='light green',activebackground='red').grid(row=7,column=1)
root.mainloop()
```
